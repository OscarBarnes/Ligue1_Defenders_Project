I will be using pandas to create tables, so importing the package will be helpful.

```python
import pandas as pd
```

I will now scrape the website FBref for the ligue 1 table, where I will look at the teams that could potentially have some transfer targetted defenders.

```python
Ligue1Table = pd.read_html('https://fbref.com/en/comps/13/Ligue-1-Stats', 
                  attrs = {"id":"results2024-2025131_overall"})[0]
```

```python
display(Ligue1Table)
```

I'll remove the attendance, top scorer, goalkeeper, and notes columns since they will not be used to look for defenders.

```python
Ligue1Table = Ligue1Table.drop(['Attendance', 'Top Team Scorer', 'Goalkeeper', 'Notes'], axis=1)
```

```python
display(Ligue1Table)
```

Two good metrics to consider are GA (goals against) and xGA (expected goals against). I will give a rank of every team based on both categories mentioned, sum the rank and then order by this sum to find the best teams ranked by both metrics. 

```python
Ligue1Table['GARk'] = Ligue1Table['GA'].rank()
```

```python
Ligue1Table['xGARk'] = Ligue1Table['xGA'].rank()
```

```python
Ligue1Table['OvrGARk'] = Ligue1Table['GARk'] + Ligue1Table['xGARk']
```

```python
display(Ligue1Table.sort_values(by = ['OvrGARk']))
```

As expected, PSG top the ranking for lowest on both metrics. Eyeing up these defenders for transfers are harder, since they play for a more profoud team, and are less likely to move, especially after the champions league win. Monaco, Lille and Nice all have a reasonable low rating, and being the teams placed 3-5 for points, this would be expected. One team that has ranked significantly lower than their position is Marseille, having conceded the 9th most goals, despite finishing 2nd. Another team that ranked noticably different is Toulouse, however their defensive rank is much higher than their position in the table.
